Check for Null values

```python
retail_data.isnull().sum()
```

Removing Missing/NaN Data from dataset

```python
retail_data[retail_data['Item_Weight'].isnull()].head()
```

Replace Item Weight by mean of Item Type.

```python
item_wt_type = retail_data.groupby('Item_Type')['Item_Weight'].mean()
item_wt_type
```

```python
retail_data['Item_Weight'].mean()
```

There is no major difference between Item_Type wise mean and mean of total Item_Weight. So lets replace NaN values of Item_Weight by mean of total Item_Weight.

```python
retail_data['Item_Weight'] = retail_data['Item_Weight'].fillna(retail_data['Item_Weight'].mean())
```

```python
retail_data[retail_data['Item_Weight'].isnull()].head()
```

All missing Item Weight values are now replaced by Mean of all item weights.

```python
missing_outlet_size = retail_data[retail_data['Outlet_Size'].isnull()]
```

```python
retail_data[retail_data['Outlet_Type'] == 'Grocery Store'].head()
```

No more missing values in Outlet_Size

# **7. Data Post-Profiling**
---

- This section is emphasised on getting a report about the data after the data manipulation.

- You may end up observing some new changes, so keep it under check and make right observations.

```python
retail_data['Item_Fat_Content'].value_counts().plot(kind='bar')
plt.show()
```

Looks like Item Fat Content values are not in same naming convention. E.g. Low Fat is represented by LF anf low fat. Same case with Regular and reg. Lets bring it to same naming convention. 

```python
retail_data['Item_Fat_Content'] = retail_data['Item_Fat_Content'].map({'LF':'Low Fat', 'low fat':'Low Fat', 'reg':'Regular'})
```

```python
retail_data['Item_Fat_Content'].value_counts().plot(kind='bar')
plt.show()
```

Now all valueas are in same naming convention. map() is used to perform this task.

## 8.1. Univariate Analysis

### 8.1.1. Maximum Sold Item.

### What all Item Type are sold and Which are top 3?

<a name = Section812></a>
### 8.1.2. Maximum Revenue Generating Outlet Size.

### What is Outlet Size proportion in given dataset?

```python
space= np.ones(3)/21
retail_data['Outlet_Size'].value_counts().plot(kind='pie', figsize=(15,7), fontsize=13, explode=space,legend=True,autopct='%3.1f%%',shadow=True)
plt.ylabel('Outlet Size')
plt.title('Donut Plot showing the proportion of each Outlet Size')
plt.show()
```

More than 55% outlets are Small in size. 

<a name = Section813></a>
### 8.1.3. Store Opening Trend.

### How many Stores are opened each year?

```python
retail_data['Year'] = retail_data['Outlet_Establishment_Year'].dt.year
retail_data.head()
```

```python
year_grpby_count = pd.DataFrame(retail_data.groupby('Year', as_index=True)['Outlet_Identifier'].nunique())
year_grpby_count.plot(kind='bar', figsize=(15,7), fontsize=13, color='black',yticks=np.arange(0,3,1))
plt.xlabel('Year of Establishment')
plt.ylabel('Store Count')
plt.savefig('Store to Year.png')
plt.show()
```

Organization has opened 2 stores in 1985. After that One Store Store is opened in any given year. Also no new stores are opened after 2009.

<a name = Section814></a>
### 8.1.4. Sales Trend.

### What is trend of sales?

```python
bins = [0,1000,2000,3000,4000,5000,6000,7000,8000,9000,10000,11000,12000,13000,14000]
retail_data['Item_Outlet_Sales'].plot(kind='hist', bins=bins, figsize=(15,7), fontsize=13, color='blue',xticks=bins)
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.title('Histogram distribution for Sales Data')
plt.show()
```

Sales looks positively Skewed. 75% of Sales data is in the range of 0 to 3500.
